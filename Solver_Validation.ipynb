```python
from Hamiltonian import Hamiltonian
from Validation import Validate
from Visualization import Visualize
from Potentials import *
```

```python
def ISW_1D(x, L):
    """1D infinite square well potential."""
    V = np.zeros_like(x)  # zero inside the well
    return V

a0_m = 5.29177210903e-11
L = 1e-9/a0_m   
H = Hamiltonian(N=500, L=L, potential_func=ISW_1D,
                ndim=1, num_states=10)

energies,eigenstates = H.solve()
print("Eigenvalues (Hartree):", energies)
print("Eigenvalues (ev):", energies*27.211386)
```

```python
if __name__ == "__main__":
    
    # --- Global Parameters ---
    N_1D = 401         # Grid points for 1D
    N_2D = 101         # Grid points for 2D (101x101)
    N_3D = 41          # Grid points for 3D (41x41x41)
    
    L_box = 10.0       # Box length (Bohr)
    L_qho = 30.0       # Box length for QHO (Bohr)
    k_au = 1.0         # Spring constant (a.u.)
    
    num_states_to_find = 6  # How many states to solve for
    num_states_to_plot = 4  # How many of those to plot
    
    # --- 1. 1D Box ---
    test_1d_box = Validate(test_case='box_1d', N=N_1D, L=L_box,
                                   num_states=num_states_to_find)
    test_1d_box.valid_solve()
    test_1d_box.plot_test(states_to_plot=num_states_to_plot)

    # --- 2. 1D QHO ---
    test_1d_qho = Validate(test_case='qho_1d', N=N_1D, L=L_qho,
                                   num_states=num_states_to_find, k=k_au)
    test_1d_qho.valid_solve()
    test_1d_qho.plot_test(states_to_plot=num_states_to_plot)

    # --- 3. 2D Box ---
    test_2d_box = Validate(test_case='box_2d', N=N_2D, L=L_box,
                                   num_states=num_states_to_find)
    test_2d_box.valid_solve()
    test_2d_box.plot_test(states_to_plot=num_states_to_plot)
    
    # --- 4. 2D QHO ---
    test_2d_qho = Validate(test_case='qho_2d', N=N_2D, L=L_qho,
                                   num_states=num_states_to_find, k=k_au)
    test_2d_qho.valid_solve()
    test_2d_qho.plot_test(states_to_plot=num_states_to_plot)

    # --- 5. 3D Box ---
    test_3d_box = Validate(test_case='box_3d', N=N_3D, L=L_box,
                                   num_states=num_states_to_find)
    test_3d_box.valid_solve()
    test_3d_box.plot_test(states_to_plot=num_states_to_plot)

    # --- 6. 3D QHO ---
    test_3d_qho = Validate(test_case='qho_3d', N=N_3D, L=L_qho,
                                   num_states=num_states_to_find, k=k_au)
    test_3d_qho.valid_solve()
    test_3d_qho.plot_test(states_to_plot=num_states_to_plot)

    print("\n" + "="*70 + "\nALL VALIDATION TESTS COMPLETE.\n" + "="*70)
```
